```python
import numpy as np
import pandas as pd
from sklearn import ensemble
from sklearn.linear_model import LogisticRegression
import os

from assistments_workbench.config_reader import config
from assistments_workbench.models.model import classification_model
```

```python
data = pd.read_csv(os.path.join(config.get('localfiles', 'data_path'), 'arrs_data.csv'))
```

```python
data['on_grade'] = np.nan
data['grade_diff_2'] = np.nan
```

```python
data.describe()
```

```python
data = data.drop_duplicates(['id', 'prereq_skill_id','prereq_easiness'], keep='first')
```

```python
data.describe()
```

```python
pfa_data = data[['skill_id', 'easiness', 'right', 'wrong', 'correct']]
```

```python
pfa_data.describe()
```

```python
label = 'correct'
all_features = pfa_data.columns.tolist()
all_features.remove(label)
predictors = all_features
categorical_features = ['skill_id']
k = 5
```

```python
model = LogisticRegression()
classification_model(model, pfa_data, predictors, label, categorical_features, k)
```

```python
grade_diff_grouped = data.groupby('grade_diff')
```

```python
for name, group in grade_diff_grouped:
    if name == 0:
        # on grade
        data['on_grade'] = 1
        data['grade_diff_2'] = 1
    else:
        data['on_grade'] = 0
        if name > 0:
            # above grade
             data['grade_diff_2'] = 0
        elif name == -1:
            # one year ago
            data['grade_diff_2'] = 2
        else:
            # longer ago
            data['grade_diff_2'] = 3
```

```python
label = 'correct'
all_features = data.columns.tolist()
all_features.remove(label)
all_features.remove('id')
all_features.remove('grade_diff')
# all_features.remove('right')
# all_features.remove('wrong')
predictors = all_features
categorical_features = ['student_class_id', 'adaptive_mode', 'mastered', 'skill_id', 'prereq_skill_id', 'on_grade', 'grade_diff_2']
```

```python
# model = ensemble.GradientBoostingClassifier()
model = LogisticRegression()
```

```python
classification_model(model, data, predictors, label, categorical_features, k)
```
